```python
import requests as rq
import os
import re
import requests
import shutil
import glob
import csv
from bs4 import BeautifulSoup
import datetime
import numpy as np
import io
import timeit
import pandas as pd
from six.moves import urllib
```

```python
# download data from 2017/6/15 to 2020/07
date = pd.date_range(start = '6/15/2017',periods = 60,freq = 'MS')
for i in range(len(date)):
  nl_response = rq.get('https://data.tainan.gov.tw/dataset/'+str(date[i].year-1911)+'-'+str(date[i].month)+'-df-mosquito-density') 
  soup = BeautifulSoup(nl_response.text, "lxml")
  result = soup.find_all('a', class_='resource-url-analytics')
  for results in result:
    url = results.get("href")
    urllib.request.urlretrieve(url,os.path.split(url)[1])
    print('下載完成'+os.path.split(url)[1])
date = pd.date_range(start = '12/15/2016',periods = 15,freq = 'MS')
for i in range(len(date)):
  nl_response = rq.get('https://data.tainan.gov.tw/dataset/'+str(date[i].year-1911)+'-'+str(date[i].month)) 
  soup = BeautifulSoup(nl_response.text, "lxml")
  result = soup.find_all('a', class_='resource-url-analytics')
  for results in result:
    url = results.get("href")
    urllib.request.urlretrieve(url,os.path.split(url)[1])
    print('下載完成'+os.path.split(url)[1])
```

```python
# download missing data
missing = ['https://data.tainan.gov.tw/dataset/104-df-mosquito-density','https://data.tainan.gov.tw/dataset/105-df-mosquito-density','https://data.tainan.gov.tw/dataset/109-2','https://data.tainan.gov.tw/dataset/109-4']
for urls in missing:
  nl_response = rq.get(urls) 
  soup = BeautifulSoup(nl_response.text, "lxml")
  result = soup.find_all('a', class_='resource-url-analytics')
  for results in result:
    url = results.get("href")
    urllib.request.urlretrieve(url,os.path.split(url)[1])
    print('下載完成'+os.path.split(url)[1])
```

```python
# move downloaded file to a folder
cwd = os.getcwd()
path_csv=glob.glob(os.path.join(cwd,'*.csv'))
for file in path_csv:
    datapath = os.path.join(cwd,'病媒蚊密度調查')
    if not os.path.isdir(datapath):
        os.mkdir(datapath)
    shutil.move(file, datapath)
```
